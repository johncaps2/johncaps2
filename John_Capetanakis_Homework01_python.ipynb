# Problem 1
(5 points)

Write a function for the following equation:

$$ y(x) = 0.9 x^4 - 12x^2 -5x $$

The input to the function is x and it returns y.

1. Test that your function returns the correct values for $y(-3)$ and $y(5)$.
1. Use the function to make a plot of $y(x)$ for $-4 \le x \le 4$.

```python
# Your answer here

import numpy as np
import matplotlib.pyplot as plt
import math

# Define the function
def equation_f(x):
  y = 0.9 * x**4 - 12 * x**2 - 5 * x
  return y

#Now that we defined it, we need to test it for our y-values
y_negative_3 = equation_f(-3)
y_5 = equation_f(5)


# Supporting print statements to test the values
# Assert the two separate claims to state different numbers and prove our claim
assert math.isclose(y_negative_3, -20.099999999999994, abs_tol=1.e-3)
assert math.isclose(y_5, 237.5, abs_tol=1.e-3)
print(f"For x=-3, y={y_negative_3}")
print(f"For x=5, y={y_5}")
```

```python
#Now, we plot y(x)

x_values = np.linspace(-4,4,100)
y_values = equation_f(x_values)

#Code to plot our plot, and extras to define the graph and its parameters
plt.figure(figsize=(10,10))
plt.plot(x_values, y_values, 'y+', label='Y(x) = 0.9 * x**4 - 12 * x**2 - 5 * x')
plt.scatter([-3,5], [y_negative_3, y_5], color='red', label='Test Points')
plt.title('Plot of y(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

# Problem 2
(5 points)

Write a function that converts polar coordinates to cartesian coordinates.  The function takes $r,\theta$ and returns $x$ and $y$, where $\theta$ is in radians.

1. Test that your function returns the correct values for $r=1,\theta=0$ and $r=1,\theta=\pi$.
1. Use the function to make a plot of the $x$ and $y$ values that correspond to:

$$ r: 0 \le r \le 1$$
$$ \theta: 0 \le \theta \le 2\pi $$

You should first create the $r$ and $\theta$ arrays, using 100 elements in each array.

3. Recreate the figure above using the plt.scatter command, and color code each point by its index in the array.


```python
# Your answer here

import numpy as np
import matplotlib.pyplot as plt

def polar_to_cartesian_coord(r, theta):
  x = r * np.cos(theta)
  y = r * np.sin(theta)
  return x, y

# Now, we teset the function with our given values for r and theta;
# First, we want to input r = 1, and theta = 0; then, we will do
# r = 1, theta = pi

x1, y1 = polar_to_cartesian_coord(1,0)
x2, y2 = polar_to_cartesian_coord(1, np.pi)

print(f"For r=1, theta=0: x={x1}, y={y1}")
# Statement to print the values for r and theta at coordinates (x1,y1)

print(f"For r=1, theta=pi: x={x2}, y={y2}")
# Statement to print the values for r and theta at coordinates (x2,y2)

# Now, we want to plot our x and y values for r: 0 <= r <= 1, and for
# theta, which is between 0 <= r <= 2*pi

r_values = np.linspace(0,1,100)
theta_values = np.linspace(0,2 * np.pi, 100)

x_values, y_values = np.meshgrid(r_values, theta_values)
cartesian_values = polar_to_cartesian_coord(x_values, y_values)

plt.figure(figsize=(10,10))
plt.polar(theta_values, r_values, 'k*', label='Polar Coordinates')
plt.scatter(cartesian_values[0], cartesian_values[1], label='Cartesian Coordinates')
plt.title('Conversion or Polar Coordinates to Cartesian Coordinates')
plt.legend()
plt.show()
```

# Problem 3

Write a function called guess_number that takes in a number between one and 10 and compares the input with a random integer (using np.random.randint(1,high=10) to select the random number).  



*   Call a different random number each time the function is called.
*   Make sure the input number is between 1 and 10.  If not, print a warning message.
*   If the input number and the random number are the same, print "You win!".
*   If the input number is 7, print "Lucky 7s always win!".
*   Otherwise, print a message to let the user know their guess was wrong, and let them know what the random number was.
*   Test your function to make sure it works.



```python
# Your function should go here
def guess_number(myguess):
  # Here, we want to assess that our input number is in between 1 and 10
  if not 1 <= myguess <= 10:
    print("Warning! Enter a number between 1 and 10 and try again")
    return
  # Now, we want to generate a numbet that is actually between 1 and 10
  random_number = np.random.randint(1, 10)

  # Compare our input number with the random number selected
  if myguess == random_number:
    print("Congratulations! You win!")
  elif myguess == 7:
    print("Lucky 7s always win!")
  else:
    print(f"Sorry, your guess was incorrect. The random number that was chosen was {random_number}")
```

```python
# Put your test code here
guess_number(11)
guess_number(7)
guess_number(2)
```
